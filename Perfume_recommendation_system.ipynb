```python
import kagglehub
import os
import pandas as pd

# Download dataset via kagglehub
dataset_path = kagglehub.dataset_download("nandini1999/perfume-recommendation-dataset")

# Check files inside the dataset folder
print("Dataset path:", dataset_path)
print("Files:", os.listdir(dataset_path))

# Load the right CSV
csv_file_path = os.path.join(dataset_path, "final_perfume_data.csv")  # Corrected filename
perfume_df = pd.read_csv(csv_file_path)

perfume_df.head()
```

```python
for col in perfume_df.columns[:2]:
    print(perfume_df[col].value_counts())
```

```python
def text_rep(row):
  text_rep = f'''
Name: {row['Name']},
Brand: {row['Brand']},
Notes: {row['Notes']},
Description: {row['Description']}
  '''
  return text_rep
```

```python
perfume_df['text_rep'] = perfume_df.apply(text_rep, axis=1)
perfume_df.head()
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load a pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Get the dimension of the embeddings from the model
dim = model.get_sentence_embedding_dimension()
index = faiss.IndexFlatL2(dim)

# Generate embeddings for the text representations
X = np.array(model.encode(perfume_df['text_rep'].tolist()), dtype='float32')

# Add the embeddings to the FAISS index
index.add(X)

print("Embeddings generated and added to FAISS index.")
```

```python
def recommend_perfumes(query, num_recommendations=5):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding, dtype='float32'),
                                      num_recommendations)
    recommended_perfume_indices = indices[0]
    recommended_perfumes = perfume_df.iloc[recommended_perfume_indices]

    return recommended_perfumes
```

```python
import gradio as gr

def get_perfume_recommendations_gradio(query):
    recommendations = recommend_perfumes(query)
    return recommendations[['Name', 'Brand', 'Notes', 'Description']]

interface = gr.Interface(
    fn=get_perfume_recommendations_gradio,
    inputs=gr.Textbox(label="Enter a perfume query (e.g., 'sweet vanilla scent')"),
    outputs=gr.Dataframe(label="Recommended Perfumes")
)

interface.launch(debug=True, share = True)
```

```python
# Save the Sentence Transformer model
model_save_path = "./sentence_transformer_model"
model.save(model_save_path)

print(f"Sentence Transformer model saved to: {model_save_path}")
```

```python
# Save the FAISS index
index_save_path = "./faiss_index.bin"
faiss.write_index(index, index_save_path)

print(f"FAISS index saved to: {index_save_path}")
```

```python
import gradio as gr
import faiss
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
import os # Import os to join paths

# --- Load the necessary components ---

# Assuming your data file, saved model folder, and saved index file
# are in the same directory as this script, or you provide correct paths.
data_file_path = "final_perfume_data.csv" # Replace with the actual path to your data
model_load_path = "./sentence_transformer_model" # Replace with the actual path to your saved model folder
index_load_path = "./faiss_index.bin" # Replace with the actual path to your saved index file

# Load your perfume data
try:
    perfume_df = pd.read_csv(data_file_path, encoding='latin-1')
except UnicodeDecodeError:
    perfume_df = pd.read_csv(data_file_path, encoding='cp1252')

# Load the saved Sentence Transformer model
model = SentenceTransformer(model_load_path)

# Load the saved FAISS index
index = faiss.read_index(index_load_path)

# --- Define the recommendation function (similar to before) ---

def recommend_perfumes(query, num_recommendations=5):
    # Encode the query using the loaded model
    query_embedding = model.encode([query])

    # Perform a similarity search in the loaded FAISS index
    distances, indices = index.search(np.array(query_embedding, dtype='float32'), num_recommendations)

    # Get the recommended perfume details from the loaded DataFrame
    recommended_perfume_indices = indices[0]
    recommended_perfumes = perfume_df.iloc[recommended_perfume_indices]

    return recommended_perfumes

# --- Define the Gradio prediction function (similar to before) ---

def get_perfume_recommendations_gradio(query):
    recommendations = recommend_perfumes(query)
    return recommendations[['Name', 'Brand', 'Notes', 'Description']]

# --- Create and launch the Gradio interface (similar to before) ---

interface = gr.Interface(
    fn=get_perfume_recommendations_gradio,
    inputs=gr.Textbox(label="Enter a perfume query (e.g., 'sweet vanilla scent')"),
    outputs=gr.Dataframe(label="Recommended Perfumes")
)

if __name__ == "__main__":
    interface.launch() # You might remove debug=True for production
```
